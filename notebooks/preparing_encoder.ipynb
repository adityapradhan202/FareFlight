```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from joblib import dump
```

```python
df = pd.read_csv("../data_folder/processed_data.csv")
df.head()
```

```python
df = df.drop("Flight_code", axis=1)
df.head()
```

```python
df.sample(10)
```

```python
numeric_cols = df.select_dtypes(include="number")
non_numeric_cols = df.select_dtypes(include="object")
```

```python
numeric_cols.head()
```

```python
non_numeric_cols.head()
```

```python
encoder = OneHotEncoder(categories="auto", # determine categories automatically from training data
                        handle_unknown="ignore", # makes other one hot cat zero
                        drop="first", # drop the first to avoid inverted duplicates
                        dtype=int) # dtype should be int
encoder.fit(non_numeric_cols)
```

```python
dump(value=encoder, filename="../models/cat_encoder.joblib")
print("Encoder has been saved!")
```
